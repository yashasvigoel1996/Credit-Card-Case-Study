```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
customer_acqusition = pd.read_csv("Customer Acqusition.csv")
customer_acqusition.head()
```

```python
# Solution 1 (a). In case age is less than 18, replace it with mean of age values.
customer_acqusition["Age"] = np.where((customer_acqusition.Age<18),customer_acqusition.Age.mean(),customer_acqusition.Age) 
customer_acqusition
```

```python
customer_spend=pd.read_csv("spend.csv")
customer_spend.head()
```

```python
customer_acqusition_spend = pd.merge(left=customer_acqusition,right=customer_spend,left_on=["Customer"],
                                     right_on=["Customer"],how="right",indicator=True)
customer_acqusition_spend.head()
```

```python
# solution 1 (b). In case spend amount is more than the limit, replace it with 50% of that customer’s limit.
customer_acqusition_spend["Amount"]=np.where((customer_acqusition_spend.Amount>customer_acqusition_spend.Limit),
                                             (customer_acqusition_spend.Limit/2),customer_acqusition_spend.Amount)
customer_acqusition_spend.head()
```

```python
customer_repayment = pd.read_csv("Repayment.csv").iloc[:,range(0,4)]
customer_repayment.head()
```

```python
#solution 1 (c). Incase the repayment amount is more than the limit, replace the repayment with the limit.
customer_acqusition_repayment=pd.merge(left=customer_acqusition,right=customer_repayment,left_on=["Customer"],
                                       right_on=["Customer"],how="right",indicator=True)
customer_acqusition_repayment.head()
```

```python
customer_acqusition_repayment["Amount"]=np.where((customer_acqusition_repayment.Amount>customer_acqusition_spend.Limit),
                                             (customer_acqusition_spend.Limit),customer_acqusition_spend.Amount)
customer_acqusition_repayment.head()
```

```python
# 2. From the above dataset create the following summaries:
# Solution a. How many distinct customers exist?
customer_acqusition.Customer.nunique()
```

```python
#Solution b. How many distinct categories exist?
customer_acqusition_spend.Type.nunique()
```

```python
#Soultion c. What is the average monthly spend by customers?
customer_acqusition_spend["Transaction_date"]=pd.to_datetime(customer_acqusition_spend.Month,format="%d-%b-%y")
avg_monthly_spend = customer_acqusition_spend.copy()
avg_monthly_spend["unique_identify"]=avg_monthly_spend.Transaction_date.dt.month.astype(str)+"-"+avg_monthly_spend.Transaction_date.dt.year.astype(str)
avg_monthly_spend=avg_monthly_spend.groupby(by=["Customer","unique_identify"])[["Amount"]].mean().add_prefix("Average_monthly_spend_")
```

```python
avg_monthly_spend.head()
```

```python
#Solution d. What is the average monthly repayment by customers?
customer_acqusition_repayment["Transaction_date"]=pd.to_datetime(customer_acqusition_repayment.Month,format="%d-%b-%y")
avg_monthly_repayment = customer_acqusition_repayment.copy()
avg_monthly_repayment["unique_identify"]=avg_monthly_repayment.Transaction_date.dt.month.astype(str)+"-"+avg_monthly_repayment.Transaction_date.dt.year.astype(str)
avg_monthly_repayment=avg_monthly_repayment.groupby(by=["Customer","unique_identify"])[["Amount"]].mean().add_prefix("Average_monthly_repayment_")
avg_monthly_repayment.head()
```

```python
#Solution e.e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month?
bank_monthly_repayment=customer_acqusition_repayment.copy()
bank_monthly_repayment["unique_identify"]=bank_monthly_repayment.Transaction_date.dt.month.astype(str)+"-"+bank_monthly_repayment.Transaction_date.dt.year.astype(str)
bank_monthly_repayment=bank_monthly_repayment.groupby(by=["unique_identify"])[["Amount"]].sum().add_prefix("Bank_monthly_repayment_")
bank_monthly_repayment.index.names = ['Month']
bank_monthly_repayment.head()
```

```python
bank_monthly_spend=customer_acqusition_spend.copy()
bank_monthly_spend["unique_identify"]=bank_monthly_spend.Transaction_date.dt.month.astype(str)+"-"+bank_monthly_spend.Transaction_date.dt.year.astype(str)
bank_monthly_spend=bank_monthly_spend.groupby(by=["unique_identify"])[["Amount"]].sum().add_prefix("Bank_monthly_spend_")
bank_monthly_spend.index.names = ['Month']
bank_monthly_spend.head()
```

```python
bank_monthly_profit=pd.merge(left=bank_monthly_repayment,right=bank_monthly_spend,
                             left_on=["Month"],right_on=["Month"],how="outer",indicator=True)
bank_monthly_profit["Profit"]=bank_monthly_profit.Bank_monthly_repayment_Amount-bank_monthly_profit.Bank_monthly_spend_Amount
bank_monthly_profit=bank_monthly_profit.loc[(bank_monthly_profit["Profit"]>0),]
bank_monthly_profit["Monthly_Profit"]=((bank_monthly_profit["Profit"]*2.9)/100)
bank_monthly_profit.index.names = ['Month']
bank_monthly_profit
```

```python
# Solution (f). What are the top 5 product types?
top_five_product_types = customer_acqusition_spend.copy()
top_five_product_types=top_five_product_types.groupby(by=["Type"])[["Type"]].count().add_prefix("Count_Of")
top_five_product_types.nlargest(5,"Count_OfType")
```

```python
# Solution (g). Which city is having maximum spend?
city_max_spend=customer_acqusition_spend.copy()
city_max_spend=city_max_spend.groupby(by=["City"])[["Amount"]].sum().add_prefix("Total_")
city_max_spend.nlargest(1,"Total_Amount")
```

```python
# Solution (h). Which age group is spending more money?
age_group_spend=customer_acqusition_spend.copy()

#Dividing age into age group by below condition
# (18-25) - > adult
# (26-45) - > middle age
# (46-60) - > senior adult
# (61-80) - > senior citizen

bins=[18,26,46,61,80]
labels=["adult","middle age","senior adult","senior citizen"]

age_group_spend["age_group"]=pd.cut(age_group_spend["Age"], bins=bins, labels=labels, right=False)
age_group_spend_more=age_group_spend.groupby(by=["age_group"])[["Amount"]].sum().add_prefix("Total_spend_")
age_group_spend_more.nlargest(1,"Total_spend_Amount")
```

```python
# Solution (i). Who are the top 10 customers in terms of repayment?
top_ten_customer_repayment=customer_acqusition_repayment.copy()
top_ten_customer_repayment=top_ten_customer_repayment.groupby(by=["Customer"])[["Amount"]].sum().add_prefix("Total_repayment_")
top_ten_customer_repayment.nlargest(10,"Total_repayment_Amount")
```

```python
#Solution 3. Calculate the city wise spend on each product on yearly basis.Also include a graphical representation for the same.

city_product_yearly_spend=customer_acqusition_spend.copy()
city_product_yearly_spend["Yearly_Identifier"]=city_product_yearly_spend.Transaction_date.dt.year.astype(str)
city_product_yearly_spend=city_product_yearly_spend.groupby(by=["City","Product","Yearly_Identifier"])[["Amount"]].sum().add_prefix("Total_").unstack()
city_product_yearly_spend.plot(kind = "bar",title="city wise spend on each product on yearly basis",figsize=(20,12))
plt.show()
```

```python
# Solution 4. Create graphs for
# (a). Monthly comparison of total spends, city wise

city_monthly_comparison_spend=customer_acqusition_spend.copy()
city_monthly_comparison_spend["Monthly_Identifier"]=city_monthly_comparison_spend.Transaction_date.dt.month.astype(str)+"-"+city_monthly_comparison_spend.Transaction_date.dt.year.astype(str)
city_monthly_comparison_spend=city_monthly_comparison_spend.groupby(by=["City","Monthly_Identifier"])[["Amount"]].sum().add_prefix("Total_").unstack()
city_monthly_comparison_spend.plot(kind = "bar",title="City Wise Monthly comparison of total spends",figsize=(20,12))
plt.show()
```

```python
# solution b. Comparison of yearly spend on air tickets
yearly_spend_air_tickets=customer_acqusition_spend.copy()
yearly_spend_air_tickets=yearly_spend_air_tickets.loc[(yearly_spend_air_tickets["Type"]=="AIR TICKET"),]
yearly_spend_air_tickets["Yearly_Identifier"]=yearly_spend_air_tickets.Transaction_date.dt.year.astype(str)
yearly_spend_air_tickets=yearly_spend_air_tickets.groupby(by=["Yearly_Identifier"])[["Amount"]].sum().add_prefix("Total_").unstack()
yearly_spend_air_tickets.plot(kind = "bar",title="Yearly spend on air tickets",figsize=(14,8))
plt.xlabel("Total_Amount,Year")
plt.show()
```

```python
# Solution c. Comparison of monthly spend for each product (look for any seasonality that exists in terms of spend)
from statsmodels.tsa.seasonal import seasonal_decompose

monthly_spend_product=customer_acqusition_spend.copy()
monthly_spend_product["Monthly_Identifier"]=monthly_spend_product.Transaction_date.dt.month.astype(str)+"-"+monthly_spend_product.Transaction_date.dt.year.astype(str)
monthly_spend_product=monthly_spend_product.groupby(by=["Product","Monthly_Identifier"])[["Amount"]].sum().add_prefix("Total_").unstack()
monthly_spend_product=monthly_spend_product.transpose().reset_index()
monthly_spend_product=monthly_spend_product.set_index("Monthly_Identifier").iloc[:,range(1,4)]
monthly_spend_product['Date'] = pd.to_datetime(monthly_spend_product.index)
monthly_spend_product = monthly_spend_product.set_index('Date')
monthly_spend_product["Silver"]=monthly_spend_product.Silver.fillna(monthly_spend_product.Silver.mean())
monthly_spend_product


```

```python
# Let see if there is any seasonality in Gold Spend
result = seasonal_decompose(monthly_spend_product.Gold, model='additive', freq=1)
result.seasonal.plot()
plt.show()

#So, there is no seasonality in Gold Spend.
```

```python
# Let see if there is any seasonality in Platimum Spend
result = seasonal_decompose(monthly_spend_product.Platimum, model='additive', freq=1)
result.seasonal.plot()
plt.show()

#So, there is no seasonality in Platimum Spend.
```

```python
# Let see if there is any seasonality in Silver Spend
result = seasonal_decompose(monthly_spend_product.Silver, model='additive', freq=1)
result.seasonal.plot()
plt.show()

#So, there is no seasonality in Silver Spend.
```
